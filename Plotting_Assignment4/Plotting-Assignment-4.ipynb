14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customize the grid lines with rendering a larger grid (major grid) and a smaller grid (minor grid). Turn on the grid but turn off ticks.

image.png

```python
#@title { vertical-output: true}
a = np.random.rand(5) + 700
dates = pd.date_range("2016-10-03", periods=len(a))
plt.figure()
plt.plot(dates, a)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing stock value of Alphabet Inc.')
plt.minorticks_on()

# Customize the major grid
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Turn off the display of all ticks.
plt.tick_params(which='both', # Options for both major and minor ticks
                top='off', # turn off top ticks
                left='off', # turn off left ticks
                right='off',  # turn off right ticks
                bottom='off') # turn off bottom ticks
plt.show()
```

15. Write a Python program to create multiple plots.


image.png

```python
#@title { vertical-output: true}

plot1 = plt.subplot2grid((2, 3), (0, 0), colspan=3)
plot2 = plt.subplot2grid((2, 3), (1, 0), rowspan=1, colspan=1)
plot3 = plt.subplot2grid((2, 3), (1, 1), rowspan=1, colspan=1)
plot4 = plt.subplot2grid((2, 3), (1, 2), rowspan=1, colspan=1)

x = np.arange(1, 10)

plot2.plot(x, x**0.5)
plot2.set_title('Square Root')

plot1.plot(x, np.exp(x))
plot1.set_title('Exponent')

plot3.plot(x, x*x)
plot3.set_title('Square')

plot4.plot(x, x*2)
plot4.set_title('Linear')

plt.tight_layout()
plt.show()
```

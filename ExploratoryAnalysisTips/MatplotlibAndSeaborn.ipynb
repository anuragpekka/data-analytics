```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate values for x-axis  (1 to 9)
x = np.arange(1, 10)

# Generate values for y-axis (1 to 18)
y = np.arange(1, 19)

# Select the first 9 values from y to match x-axis length
y = y[:9]  # Slicing to get values from index 0 (inclusive) to 9 (exclusive)

# Create the line plot
plt.plot(x, y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Line Plot Title Here')
plt.show()

# Saving the plot
plt.savefig('line_plot.png')
```
